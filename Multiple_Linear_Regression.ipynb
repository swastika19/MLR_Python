Correlation Matrix and Visualisations

```python
import seaborn as sns
sns.set_style('darkgrid')
```

```python
corr = dataset2.corr()
#Plot figsize
fig, ax = plt.subplots(figsize=(12, 12))
#Generate Heat Map, allow annotations and place floats in map
mask = np.tril(corr)
sns.heatmap(corr, annot=True, mask=mask)
#Apply xticks
plt.xticks(range(len(corr.columns)), corr.columns);
#Apply yticks
plt.yticks(range(len(corr.columns)), corr.columns)
#show plot
plt.show()
```

```python
corr1 = dataset2.corr()
#Plot figsize
fig, ax = plt.subplots(figsize=(12, 12))
#Generate Heat Map, allow annotations and place floats in map
mat = np.triu(corr1)
sns.heatmap(corr1, annot=True, mask=mat)
#Apply xticks
plt.xticks(range(len(corr1.columns)), corr1.columns);
#Apply yticks
plt.yticks(range(len(corr1.columns)), corr1.columns)
#show plot
plt.show()
```

**Linear Regression Model**

```python
#Importing required Packages
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#Initialise Model
regression_model = LinearRegression()

regression_model.fit(X_train, Y_train)
```

Predicting Prices from Training Set

```python
Y_pred = regression_model.predict(X_train)
Y_pred
```

```python
rmse = mean_squared_error(Y_train, Y_pred)
r2 = r2_score(Y_train, Y_pred)
```

```python
print('Slope:' ,regression_model.coef_)
print('Intercept:', regression_model.intercept_)
print('Root mean squared error: ', rmse)
print('R2 score: ', r2)
```
